```python
import pandas as pd
products = pd.DataFrame({
"product_id": [101, 102, 103, 104, 105],
"product_name": ["Laptop", "Phone", "Tablet", "Headphones", "Smartwatch"],
"category": ["Electronics", "Electronics", "Electronics", "Accessories", "Accessories"],
"price": [800, 500, 300, 100, 200]
})
sales = pd.DataFrame({
"sale_id": [1, 2, 3, 4, 5, 6, 7],
"product_id": [101, 102, 103, 104, 101, 105, 106], # 106 not in products
"quantity": [3, 5, 2, 10, 1, 4, 2]
})
print("=== Products Dataset ===")
print(products, "\n")
print("=== Sales Dataset ===")
print(sales, "\n")
merged_inner = pd.merge(sales, products, on="product_id", how="inner")
print("=== Inner Join (Matching Records Only) ===")
print(merged_inner, "\n")
merged_outer = pd.merge(sales, products, on="product_id", how="outer")
print("=== Outer Join (All Records) ===")
print(merged_outer, "\n")
merged_inner["revenue"] = merged_inner["quantity"] * merged_inner["price"]
total_revenue = merged_inner["revenue"].sum()
average_price = merged_inner["price"].mean()
product_sales = merged_inner.groupby(["product_name", "category"]).agg(
total_quantity=("quantity", "sum"),
total_revenue=("revenue", "sum")
).reset_index()
category_sales = product_sales.groupby("category")["total_revenue"].sum().reset_index()
best_category = category_sales.loc[category_sales["total_revenue"].idxmax()]
print("=== Product-wise Sales ===")
print(product_sales, "\n")
print("=== Category-wise Revenue ===")
print(category_sales, "\n")
print(f"Total Revenue: {total_revenue}")
print(f"Average Price of Sold Products: {average_price:.2f}")
print(f"Best-Selling Category: {best_category['category']} "
f"with Revenue {best_category['total_revenue']}")
```

```python
import pandas as pd
teachers = pd.DataFrame({
"teacher_id": [1, 2, 3, 4, 5],
"teacher_name": ["Alice", "Bob", "Charlie", "David", "Eva"],
"department": ["CSE", "ECE", "CSE", "ME", "ECE"]
})
courses = pd.DataFrame({
"course_id": [101, 102, 103, 104, 105, 106, 107],
"course_name": ["DSA", "Networks", "DBMS", "Thermodynamics",
"AI", "VLSI", "OS"],
"teacher_id": [1, 2, 1, 4, 3, 2, 6] # Note: teacher_id=6 not in teachers
})
print("=== Teachers Dataset ===")
print(teachers, "\n")
print("=== Courses Dataset ===")
print(courses, "\n")
merged_inner = pd.merge(courses, teachers, on="teacher_id", how="inner")
print("=== Inner Join (Matching Teachers & Courses) ===")
print(merged_inner, "\n")
merged_left = pd.merge(courses, teachers, on="teacher_id", how="left")
print("=== Left Join (All Courses, Missing Teachers as NaN) ===")
print(merged_left, "\n")
courses_per_teacher = merged_inner.groupby(["teacher_id", "teacher_name"]).size().reset_index(name="num_courses")
average_courses = courses_per_teacher["num_courses"].mean()
dept_workload = merged_inner.groupby("department")["course_id"].count().reset_index()
dept_workload.rename(columns={"course_id": "total_courses"}, inplace=True)
print("=== Courses per Teacher ===")
print(courses_per_teacher, "\n")
print("=== Department-wise Workload ===")
print(dept_workload, "\n")
print(f"Average Courses per Teacher: {average_courses:.2f}")
```
