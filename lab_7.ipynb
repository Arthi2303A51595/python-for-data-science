```python
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset
data = pd.DataFrame({
    "Day": list(range(1, 31)),   # 30 days
    "Temperature": [28, 29, 30, 32, 31, 29, 28, 27, 30, 33,
                    34, 35, 36, 34, 33, 31, 29, 28, 27, 26,
                    28, 30, 32, 33, 34, 32, 31, 29, 28, 27]
})

# Plot
plt.figure(figsize=(10, 6))
plt.plot(data["Day"], data["Temperature"], marker='o', linestyle='-', color='b', label="Temperature")
plt.title("Daily Temperature Changes (1 Month)", fontsize=14)
plt.xlabel("Day", fontsize=12)
plt.ylabel("Temperature (°C)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()

# Find hottest and coldest days
hottest_day = data.loc[data["Temperature"].idxmax()]
coldest_day = data.loc[data["Temperature"].idxmin()]

print(f"Hottest Day: Day {hottest_day['Day']} with {hottest_day['Temperature']}°C")
print(f"Coldest Day: Day {coldest_day['Day']} with {coldest_day['Temperature']}°C")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset
sales = pd.DataFrame({
    "Quarter": ["Q1", "Q2", "Q3", "Q4"],
    "Sales": [15000, 18000, 17000, 22000]
})

# Plot
plt.figure(figsize=(8, 5))
plt.plot(sales["Quarter"], sales["Sales"], marker='o', linestyle='-', color='g', label="Sales")

# Highlight peak sales
peak = sales.loc[sales["Sales"].idxmax()]
plt.scatter(peak["Quarter"], peak["Sales"], color='red', s=100, zorder=5, label="Peak Sales")

# Titles and labels
plt.title("Quarterly Sales Performance", fontsize=14)
plt.xlabel("Quarter", fontsize=12)
plt.ylabel("Sales (in units)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
